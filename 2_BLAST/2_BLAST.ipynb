# Random sequences generation

```python
import numpy as np

np.random.seed(42)

for i in [100, 1000, 10_000]:
    with open('./sequence_{}.txt'.format(i), 'w') as f:
        f.write(''.join(np.random.choice(['A', 'G', 'C', 'T'], size=i)))
```
